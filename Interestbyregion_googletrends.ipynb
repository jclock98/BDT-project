```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='it-IT', geo='IT')
kw_list = ["sport"]
pytrends.build_payload(kw_list,timeframe = 'now 1-d' )
df = pytrends.interest_by_region(resolution='Italy', inc_low_vol=True, inc_geo_code=False)
df.sort_values(by=kw_list,ascending=False)
```
