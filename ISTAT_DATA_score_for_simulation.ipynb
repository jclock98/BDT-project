```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import pickle 
import csv
```

```python
#Loading the dataset
istatdata = pd.read_csv("DCCV_AVQ_PERSONE_07062022160135090.csv")
```

```python
#Visualizing our data
istatdata
```

```python
#Selecting 2021 data
istatdata21 = istatdata[istatdata.TIME==2021]
#Selecting percentage data
istatdata21 = istatdata21[istatdata21.MISURA_AVQ=="HSC"]
```

```python
#Creating a new dataset with selected and defined data
istatdata21 = istatdata21.copy(deep=True)
```

```python
#Now we want to transform qualitative labels into numbers to 
# have a quantitative ranking

#Defining a "multiplier" 
multiplier = []
#Defining a dictionary to map the label to a quantitative weight
ranktomultiplier = {}

ranktomultiplier['non praticano sport, né attività fisica'] = 0
ranktomultiplier["solo qualche attività fisica"] = 1
ranktomultiplier["in modo saltuario"] = 2
ranktomultiplier["in modo continuativo"] = 3

#Adding the multiplier to the dataframe
for el in istatdata21["Tipo dato"]:
    multiplier.append(ranktomultiplier[el]) 
istatdata21["multiplier"] = multiplier
```

```python
#Multiplying the value referred to each label by the defined multiplier
istatdata21["score"] = istatdata21["Value"] * istatdata21["multiplier"] 
```

```python
istatdata21
```

```python
#GROUP BY TERRITORIES AND SUMMING THE SCORE WITH THE DECIDED METRIC
level = istatdata21.groupby('Territorio')['score'].sum()
```

```python
#CONVERTING TO DATA FRAME
physical_data = level.to_frame()
#MIN-MAX NORMALIZATION ()
maxvalue = 300
minvalue = 0
physical_data["score"] = round((physical_data["score"] - minvalue) / (maxvalue - minvalue),2)
```

```python
nordlist = ["lombardia","emilia","trento","bolzano","friuli","liguria","aosta","piemonte","veneto"]
Northen_Italy = []
for x in physical_data.index:
    for y in nordlist:
        if y in x.lower():
            Northen_Italy.append(x)
```

```python
#ORDER BY SCORE DESC
physical_data = physical_data.sort_values(by="score",ascending=False)
```
